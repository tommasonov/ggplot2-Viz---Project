```r
df <- read.csv('C:/Users/worme/Desktop/R course/R-Course-HTML-Notes/R-for-Data-Science-and-Machine-Learning/Training Exercises/Capstone and Data Viz Projects/Data Visualization Project/Economist_Assignment_Data.csv')
```

```r
options(repr.plot.width=8, repr.plot.height=4)
```

```r
pl <- ggplot(df,aes(x=CPI,y=HDI,color=Region)) + geom_point(size=2)
pl
```

```r
pl <- ggplot(df,aes(x=CPI,y=HDI,color=Region)) + geom_point(shape=1,size=4)
pl
```

```r
pl <- ggplot(df,aes(x=CPI,y=HDI,color=Region)) + geom_point(shape=1,size=4) + geom_smooth(aes(group=1))
pl
```

```r
pl2 <- ggplot(df,aes(x=CPI,y=HDI,color=Region)) + geom_point(shape=1,size=4) + geom_smooth(aes(group=1),method ='lm',formula = y ~ log(x),se=FALSE,color='red')
pl2
```

```r
pl2 + geom_text(aes(label=Country))
```

```r
pointsToLabel <- c("Russia", "Venezuela", "Iraq", "Myanmar", "Sudan",
                   "Afghanistan", "Congo", "Greece", "Argentina", "Brazil",
                   "India", "Italy", "China", "South Africa", "Spane",
                   "Botswana", "Cape Verde", "Bhutan", "Rwanda", "France",
                   "United States", "Germany", "Britain", "Barbados", "Norway", "Japan",
                   "New Zealand", "Singapore")
pl3 <- pl2 + geom_text(aes(label = Country), color = "gray20", 
                data = subset(df, Country %in% pointsToLabel),check_overlap = TRUE)

pl3
```

```r
pl4 <- pl3 + theme_bw()
pl4
```

```r
pl4 <- pl4 + scale_x_continuous(name='Corruption Perceptions Index, 2011 (10=least corrupt)',limits=c(.9,10.5),breaks=1:10)
pl4
```

```r
pl4 <- pl4 + scale_y_continuous(name='Human Development Index, 2011 (1=B)',limits=c(0.2,1))
pl4
```

```r
pl4 + ggtitle("Corruption and Human development")
```
